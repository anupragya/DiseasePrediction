```python
gender_dict={'Male':0,'Female':1}
df["Gender"]=df["Gender"].map(gender_dict)
```

```python
yesno_dict={'No':0,'Yes':1}
df["Smoking"]=df["Smoking"].map(yesno_dict)
df["Alcohol"]= df["Alcohol"].map(yesno_dict)
df["FamilyHistory"]=df["FamilyHistory"].map(yesno_dict)
```

```python
df.Disease.unique()
```

```python
PA_dict={'Medium':1, 'High':2, 'Low':0}
df["PhysicalActivity"]=df["PhysicalActivity"].map(PA_dict)
```

```python
di_dict= {'Healthy':0, 'Pre-Diabetes':1, 'Hypertension':2, 'Heart Disease':3, 'Diabetes':4}

df["Disease"]=df["Disease"].map(di_dict)
```
